# 3.1 ANALYSIS OF THE USERS (Customers)

First, we are going to create a new dataframe which provide us of more information about the users, for this purpose we are going to use the function *merge* between the dataframes *single_rev1* and *single_user*, using as key variable the *Username*, but just by the customers who have make a review, in other words with an *inner join* by default.

# 3.1.1 Merge between auxiliar dataframes

```python
#Create a data frame to merge between single_user and single_rev1, just by the clients who has make a review, 
#Using an inner join function by default:
char_user= pd.merge(single_rev1, single_user, on="Username")[["Uniq_id","Score", "State", "Age", "Review"]]
```

```python
char_user.head()
```

We can see that there is a new dataframe wich is called now **char_user** with more variables added by the analysis. It has now 5 variables. Now we are going to check this new data frame:

```python
#Check the information with .info():
char_user.info()
```

Now that we have picked up the Age of the users, we can analyse what is the score that the users give according the Age:

```python
#Analyze the Age of users by Score using the functions groupby and describe:
print(char_user[char_user.Score.isin([0,1,2,3,4,5])].groupby("Score")["Age"].describe())
```

We can see that for the lowest score the mean of Age is closer 48 years old, but for all the scores we can see a similar behaviour, for example, at least the 25% of the customers of all the scores (0-5) have 35 years old. The most younger customers have 22 years old.

Now we are going to create a scatterplot to know the distribution of the score that give the users to the products, by State and Age to check what type of information can get:

# 3.1.2 Distribution of the score given by the Users of the products by State and Age

```python
#To apply the height and wide of the figure with figsize:
plt.figure(figsize=(12, 6))
#To create a scatterplot using sns, specify the axis and give color by State, here use the palette "magma":
sns.scatterplot(x='Age', y='Score', hue='State', data=char_user, palette='magma', alpha=0.7)
#To show the labels and title using plt:
plt.title('Scatter Plot by State')
plt.xlabel('Age')
plt.ylabel('Score')
#To show the legend with the name of the State "uper left":
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

We can see that there are a lot of States and also by the Age is a little difficult to understand this type of graph, for this reason we are going to create a barplot, generating the *range* of the Age to improve the visualization and also the analysis. We will remove from this analysis the variable State.

# 3.1.3 Frequency of the Users by Group of Age and Score

```python
# To define the age bins, regarding the maximum and minimum value of the age:
age_bins = [0, 22, 30, 50, 100]
#To create 4 bins with the labels:
age_labels = ["22-35", "36-47", "48-60", "61+"]

# Create a new column 'AgeGroup' based on the age bins, using .cut to segment the Age:
char_user["AgeGroup"] = pd.cut(char_user["Age"], bins=age_bins, labels=age_labels, right=False)

#Plot the count of each age group by score, adjust the size of the figure (8,6):
plt.figure(figsize=(8, 6))
#Order by the score using the function range:
sns.countplot(data=char_user, x="Score", hue="AgeGroup", order= range(0,char_user["Score"].max()+1))
#To plot the title and labels:
plt.title("Count of Users by Score and Age Group")
plt.xlabel("Score")
plt.ylabel("Count")
plt.legend(title="Age Group")
plt.show()
```

We can observe that the age group with more frequency for all the scores is the range of 61 and more. Also the behaviour for all the groups of age is similar in all the scores of qualification of the products. But, we can see that for the score 1, there is a population of more than 5500 persons with 61 years old and more. In general the score with the most highest frequency for all the range of ages is 1. Did you remember that the average of the score by the users was of 1? Here is the justification!!

There is something that does not appear in this graph, What happened with the population in the range of age between 22-35?, Why it does not appear in the bar plot?, now we are going to make a check of this group of age:

```python
#To review the population of 22 years old using groupby:
print(char_user[char_user.Age==22].groupby("Age").count())
```

Definitelly, the frequency is very small how to appear on the graph.

# 3.3.1 Polarity

Now we need to import from *textblob* from TextBlob, which is a Lexicon-based sentiment analyzer that has some predefined rules, in other words, it has word and weight dictionary, where it has some scores that help to calculate a sentence's polarity.

```python
#Import the next librarys to analyse the sentiments of the column Review:
from textblob import TextBlob
```

We are going to create a function that allow us to get the sentiment polarity:

```python
# Create a function to get sentiment polarity with def and return:
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity
```

Now we are going to apply the function over the column *Review*. We are going to store the results in a new column called *Sentiment*

```python
# Apply the function to the 'Review' column of the char_user dataframe:
char_user['Sentiment'] = char_user['Review'].apply(get_sentiment)
```

We are going to use a lambda function and an if else conditional, to classify the sentiments as: Positive, Negative or Neutral.

```python
# Classify sentiments as positive, negative, or neutral using lambda function if x>0 'Positive', if x<0 'Negative' else 'Neutral'.
char_user['Sentiment_Class'] = char_user['Sentiment'].apply(lambda x: 'Positive' if x > 0 
                                                            else ('Negative' if x < 0 
                                                                else 'Neutral'))
```

Finally, print the results with all the columns created:

```python
# Display the results of the dataframe char user that applied the lambda function for sentiment analysis:
print(char_user[['Review', 'Score','Sentiment', 'Sentiment_Class']])
```
